W2H Ratio
- **ID**: unique identifier of each person, integer
- **Waist**: measurement of participant's waist circumfrence (cm), integer
- **Hip**: measurement of participant's hip circumfrence (cm), integer
- **Gender**: male or female, string
- **Shape**: WHO metric of body shap determined by W2H ratio, string

## Data Import

```python
# Goal: Extract the data from the file

# use proper pandas function to read data from a CSV file to a DataFrame

import pandas as pd
datafile=pd.read_csv('w2h_data.csv')
datafile.head()
```

## Calculations

```python
# Goal: For each row of data calculate and store the w2h_ratio and shape.

# Step 1: Make sure the columns are in correct data types - refer to previous parts for the correct data types

# Step 2: calculate the 'W2H_Ratio' column using the 'Waist' and 'Hip' columns

# Step 3: Create the `Shape` column based on the values of the `W2H_Ratio` column - refer to previous parts for the logic
```

```python
#Step 1: Make sure the columns are in correct data types - refer to previous parts for the correct data types
datafile.dtypes  #waist and hip are integers, so we do not need to change the data types

# Step 2: calculate the 'W2H_Ratio' column using the 'Waist' and 'Hip' columns
datafile['W2h_Ratio'] = datafile['Waist']/datafile['Hip']
#datafile.head() #tested the new column W2h_ratio

# Step 3: Create the `Shape` column based on the values of the `W2H_Ratio` column - refer to previous parts for the logic
datafile.loc[(datafile.Gender == 'M') & (datafile['W2h_Ratio'] > 0.9) | (datafile.Gender == 'F') & (datafile['W2h_Ratio'] > 0.8), 'shape'] = 'Apple'
datafile.loc[(datafile.Gender == 'M') & (datafile['W2h_Ratio'] <= 0.9) | (datafile.Gender == 'F') & (datafile['W2h_Ratio'] <= 0.8), 'shape'] = 'Pear'
#datafile  #tested the shape column values
```
